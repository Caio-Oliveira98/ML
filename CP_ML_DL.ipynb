                            Introdução

O Objetivo é construir um modelo para prever qual o preço de aluguel adequado para os imóveis cadastrados na plataforma "Sexto Andar". Utilizando regressão linear e descobrindo a métrica alvo com o erro quadrático menor.

Vamos trabalhar com as seguintes variáveis:

1 Cidade - Cidade onde está localizada o ímovel
2 Área - Área em m² do imóvel
3 quartos - quantidade de quartos
4 banheiros = Quantidade de banheiros
5 garagem = Quantidade de vagas na garagem
6 Andar - Andar da casa
7 animal - Se permite ou não animal doméstico
8 Mobiliado - Se o imóvel está ou nãp mobiliado
9 Condomínio - Valor da taxa de condomínio (R$)
10 Valor Aluguel - Valor do aluguel (R$)
11 Valor IPTU - Valor do IPTU (R$)
12 Valor Seguro(R$) - Valor do seguro de incêndio
13 total (R$) - Valor total de alguém do imóvel

Iremos utilizar 1 modelo inicial utilizando a variável "área", depois incrementar a variável "Mobiliado" e depois "Números de Quartos" e comparar os resultados dos erros quadráticos médios que serão feitos nos 3 modelos, sendo o menor erro, melhor o modelo.



```python
import pandas as pd #importando a biblioteca para processamento de dados
```

```python
df = pd.read_csv('/content/aluguel.csv', sep=';') #lendo o arquivo csv, e armazena os dados em um DataFrame chamado df
```

```python
df.head(10) #visualizando o data frame
```

```python
df.describe() #analise descritiva
```

```python
#validação hold-out (utilizado para evitar overfitting, avaliação de generalização, simplicidade, tempo de execução, benchmarking)

from sklearn.model_selection import train_test_split
```

```python
X = df[["area"]] #variável de entrada área
y = df[["valor_total"]] #variável resposta
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pip install scikit-learn #instalando as bibliotecas necessárias
```

```python
from sklearn.linear_model import LinearRegression #importando a regressão linear
from sklearn.metrics import mean_squared_error #importando a métrica EQM

model = LinearRegression() #criando o modelo
model.fit(X_train, y_train) #treinando o modelo
```

```python
#Fazendo previsões
y_pred = model.predict(X_test)

#Calcular O EQM
eqm = mean_squared_error(y_test, y_pred)
print("EQM:", eqm)
```

```python
#Ajustando o modelo para incluir a coluna Mobiliado
```

```python
x2 = df[["area", "mobiliaria"]]#variável de entrada área
y = df[["valor_total"]]#variável resposta
#Dividir os dados de teste e treino novamente
x2_train, x2_test, y_train, y_test = train_test_split(x2, y, test_size=0.3, random_state=42)

#ajustando o modelo
model2 = LinearRegression()
model2.fit(x2_train, y_train)

#fazer previsões
y_pred2 = model2.predict(x2_test)

#calcular o EQM com a mobiliária
eqm2 = mean_squared_error(y_test, y_pred2)
print("EQM:", eqm2)
```

```python
#Incluir "Quartos" no modelo e pedir pra ele prever o erro quadrático

x3 = df[["area", "quartos", "mobiliaria"]]#variável de entrada área
y = df[["valor_total"]]#variável resposta
#dividir os dados de teste e treino novamente
x3_train, x3_test, y_train, y_test = train_test_split(x3, y, test_size=0.3, random_state=42)

#ajustando o modelo
model3 = LinearRegression()
model3.fit(x3_train, y_train)

#fazer previsões
y_pred3 = model3.predict(x3_test)

#calcular o EQM com a mobiliária
eqm3 = mean_squared_error(y_test, y_pred3)
print("EQM:", eqm3)
```

                                Conclusão

De acordo com a análise feita após ajustar o modelo, fica claro que o modelo que inclui tanto os "Quartos" como "Mobiliária" além da "Área" teve o menor erro quadrático médio, se tornando o mais apto a ser utilizado dentro desses 3 modelos criados.
