## Implementação de um Perceptron simples

```python
# Definir a função de agregação de informação (soma poderada)
```

```python
x = [1, 2, 3]
w = [0.1, 0.2, 0.3]
b = 0
u = 0
```

```python
def soma(x, w, b):
  u = 0
  for i in range(len(x)):
    u += x[i]*w[i]
  u += b
  return u
```

```python
print(soma(x, w, b))
```

```python
# Definir a função de ativação: FUNÇÃO DEGRAU
def funcao_degrau(soma):
  if soma >= 1:
    return 1
  return 0

```

```python
sum = soma(x, w, b)
y = funcao_degrau(sum)
print(f"Para as entradas X = {x} e pesos W={w}, o perceptron foi {y}")
```
